```python
# Напишите цикл, который собирает содержимое файлов папки data в единый датафрейм data.
# Сколько строк в датафрейме data?
import pandas as pd


# desicions 1 
paths = []
for root, dirs, files in os.walk('data'): 
    paths += [
        os.path.join(root, filename)
        for filename in files
        if filename.split('.')[-1] == 'txt'
    ]

columns = ['userId', 'movieId', 'rating', 'timestamp']
data = pd.DataFrame(columns = columns)
for filename in paths:
    temp = pd.read_csv(filename, names = columns)
    data = pd.concat([data, temp])
    
len(data)


# desicions 2
paths = []
columns = ['userId', 'movieId', 'rating', 'timestamp']

data = pd.DataFrame(columns = columns)
for root, dirs, files in os.walk('data'):
    for filename in files:
        if filename.split('.')[-1] == 'txt':
            temp = pd.read_csv(os.path.join(root, filename), names = columns)
            data = pd.concat([data, temp])
            
len(data)
    
```
